### Machine Learning

```python
#import library
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
#pembuatan variabel dummy untuk variabel OFFENSE_CODE_GROUP
encoded1 = pd.get_dummies(df['OFFENSE_CODE_GROUP'], prefix='OCG')
encoded1.head()
```

```python
#pembuatan variabel dummy untuk variabel DISTRICT
encoded2=pd.get_dummies(df['DISTRICT'], prefix='district')
encoded2.head()
```

```python
#mengubah variabel nama hari menjadi angka 
day=df['OCCURRED_ON_DATE'].dt.dayofweek
#menggabungkan seluruh variabel X
X=encoded1.join([encoded2,day,df['MONTH']])
X.head()
```

```python
# membuat variabel Y
y=df['HOUR']
```

```python
# Membagi data menjadi train data dan test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# membuat klasifikasi
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
#memprediksi klasisikasi pada data test
y_pred=clf.predict(X_test)
```

```python
#Akurasi dalam satu percobaan sample data
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# akurasi dengan cross validation sebanyak 10 kali
from sklearn.model_selection import cross_val_score
cross_val_score(clf, X_train, y_train, cv=10)
```
